### Фильтр Калмана и алгоритм Метрополиса-Гастингса.

Этот ноутбук посвящен фильтру Калмана и алгоритму М.-Г., которые лежат в основе state space модели VAR. Я для себя хочу разобраться, как работают оба алгоритма.

### Фильтр Калмана.

[Источник мудрости](https://habr.com/ru/post/166693/)

Пусть у нас есть радиоуправляемая машина, координата $x_{k+1}$ которой зависит от предыдущей координаты $x_k$, некоторой переменной $u_k$ и случайной ошибки $\xi_k$. При этом у нас есть сенсор, который измеряет текущую координату с некоторой ошибкой $\nu_k$:

$$
\begin{cases}
    x_{k+1} = x_k + u_k + \xi_k \\
    z_k = x_k + \nu_k
\end{cases}
$$

Задача: зная $z_k$, найти оптимальное приближение $x_k^{\textrm{opt}}$ для $x_k$.

Будем рассуждать по индукции. Пусть нам известно первое уравнение из системы. Представим, что на шаге $k$ мы уже нашли отфильтрованное значение $x_k^{\textrm{opt}}$, тогда мы знаем, что $E(z_{k+1}) = x_k^{\textrm{opt}} + u_k$, и что мы получим от сенсора $z_{k+1}$.

Идея в том, что мы можем присвоить разные веса Калмана $K$ показанию сенсора и нашему прогнозу через матожидание:

$$
x^{\textrm{opt}}_{k+1} = K \cdot z_{k+1} + (1 - K) \cdot (x^{\textrm{opt}}_{k} + u_k)
$$

На самом деле, коэффициент Калмана зависит от шага итерации, но не будем пока усложнять. Мы хотим подобрать его так, чтобы минимизировать ошибку

$$
e_{k+1} = x_{k+1} - x^{\textrm{opt}}_{k+1}
$$

Выведем новое уравнение для ошибки, развернув, а потом посокращав все лишнее:

$$
\begin{gather}
    e_{k+1} = x_{k+1} - x^{\textrm{opt}}_{k+1} = \\[5pt]
    = x_{k+1} - K \cdot z_{k+1} - (1 - K) \cdot (x^{\textrm{opt}}_{k} + u_k) = \\[5pt]
    = x_k + u_k + \xi_k - K \cdot (x_k + u_k + \xi_k + \nu_k ) - (1 - K) \cdot (x^{\textrm{opt}}_{k+1} + u_k) = \\[5pt]
     = (1 - K) \cdot (x_k - (x^{\textrm{opt}}_{k} + \xi_k) - K \cdot \nu_k = \\[5pt]
     = \mathbf{(1 - K) \cdot (e_k + \xi_k) - K \cdot \nu_{k+1}}
\end{gather}
$$

Минимизируем квадрат ошибки (я так понимаю, квадрат удобно минимизировать и к тому же матожидание ошибки - ноль, поэтому по сути минимизируем дисперсию ошибки):

$$
\begin{gathered}
    E(e^2_{k+1}) \rightarrow min \\
    \Leftrightarrow E(e^2_{k+1}) = Var(e_{k+1}) = (1 - K)^2 \cdot (Var(e_{k+1}) + \sigma^2_{\xi}) + K^2 \sigma^2_{\nu} \\ = 
    = (1 - K)^2 \cdot (E(e^2_{k+1}) + \sigma^2_{\xi}) + K^2 \sigma^2_{\nu} 
\end{gathered}
$$

Отсюда получим через производную:

$$
K_{k+1} = \frac{Ee^2_k + \sigma^2_{\xi}}{Ee^2_k + \sigma^2_{\xi} + \sigma^2_{\nu}}
$$

На самом деле, получили вполне ожидаемый результат: веса при показаниях сенсора и оптимальном прогнозе зависят от соотношения случайных ошибок сенсора и реального процесса.

Теперь посмотрим на простой пример в коде.

```python
#  number of samples
N=100 
# acceleration
a=0.1 

sigmaPsi=5
sigmaEta=50;

k=np.arange(N)
x=k
x[0] = 0
z = np.empty((N,))
z[0] = x[0] + np.random.normal(loc=0, scale=sigmaEta)

for t in range(N-1):
    x[t+1] = x[t] + a * t+ np.random.normal(loc=0, scale=sigmaPsi, size=(1,))
    z[t+1] = x[t+1] + np.random.normal(loc=0, scale=sigmaEta, size=(1,))
    
# kalman filter
K = np.empty((N,))
xOpt = np.empty((N,))
eOpt = np.empty((N,))

xOpt[0]=z[0]
eOpt[0]=sigmaEta

for t in range(N-1):
    # eOpt(t) is a square root of the error dispersion (variance). 
    # It's not a random variable. 
    eOpt[t+1]=np.sqrt((sigmaEta**2)*(eOpt[t]**2+sigmaPsi**2)/(sigmaEta**2+eOpt[t]**2+sigmaPsi**2))
    K[t+1]=(eOpt[t+1]**2)/sigmaEta**2
    xOpt[t+1]=(xOpt[t]+a*t)*(1-K[t+1])+K[t+1]*z[t+1]
```

```python
d = pd.DataFrame({'K': K,
                  'Kalman': xOpt,
                  'sensor': z,
                  'true label': x})
d.plot(figsize=(11, 5), grid=True);
```

```python
d.K.plot(ylim=(-0.5, 1), grid=True, title='Коэффициент Калмана', style='.');
```
