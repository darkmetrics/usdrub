### 2. Модель.

Традиционно, существует 2 подхода к моделированию коинтегрированных переменных:
1. **Методология Энгла-Грейнджера** - двухшаговая регрессия: <br>
    1.1. Сначала оценивается обычная линейная регрессия для коинтегрированных переменных.<br>
    1.2. Остатки из коинтеграционного соотношения проверяются на единичный корень, после чего включаются в VAR-модель с переменными, преобразованными к стационарному виду. Остатки из регрессии в данном случае играют роль Error Correction Term, коэффициент при них в VAR - роль speed adjustment coefficient. <br>
2. **Методология Йохансена** предлагает:<br>
    2.1. Провести trace test или maximum eigenvalue тест для выявления количества коинтеграционных соотношений в данных.<br>
    2.2. Оценить VAR сразу с остатками коинтеграционного уравнения внутри, воспользовавшись данными из теста.<br>
    
У методологии Энгла-Грейнджера есть ряд общеизвестных недостатков, зафиксированных в научной литературе:
1. Число коинтеграционных соотношений в системе должен задать исследователь, зачастую это одно соотношение. От этих решений очень сильно зависит итоговый результат.
2. Так как процедура двухшаговая, ошибки спецификации из первого этапа попадают в модель на втором этапе и накладываются на потенциальные ошибки второго шага.
3. На двух этапах параметры оцениваются по отдельности, а не совместно, что также влияет на значения полученных оценок.

В данной работе мы будем использовать оба подхода: тест на коинтеграцию проведем по методологии Йохансена, а саму модель будем строить по подходу Энгла-Грейнджера, что обусловлено особенностями state-space моделей. Наша модель - это state-space Vector Error Correction Model.  

```python
# напишем простенький класс для представления результатов Johansen test
# в statsmodels реализован сам тест, но почему-то нет нормальной выдачи

class JohansenResults:
    """
    Returns Johansen test results from statsmodels
    in a pretty format.
    """
    def __init__(self, TestResults):
        self.TestResults = TestResults
    
    @property
    def max_eigenvalue(self)->pd.DataFrame:
        """Results of max eigenvalue test"""
        max_eig_obs = self.TestResults.max_eig_stat
        max_eig_crit = self.TestResults.max_eig_stat_crit_vals
        test_data = np.concatenate((max_eig_obs.reshape(-1, 1),
                                   max_eig_crit),
                                   axis=1)
        result = pd.DataFrame(data=test_data,
                            columns=['Observed', '10%', '5%', '1%'])
        result.index = np.arange(1, len(result)+1)
        result.index.name = 'Rank'
        return result.sort_index(ascending=False).round(2)
    
    @property
    def trace(self)->pd.DataFrame:
        """Results of trace test"""
        trace_obs = self.TestResults.trace_stat
        trace_crit = self.TestResults.trace_stat_crit_vals
        test_data = np.concatenate((trace_obs.reshape(-1, 1),
                                   trace_crit),
                                   axis=1)
        result = pd.DataFrame(data=test_data,
                            columns=['Observed', '10%', '5%', '1%'])
        result.index = np.arange(1, len(result)+1)
        result.index.name = 'Rank'
        return result.sort_index(ascending=False).round(2)
    
```

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
test_data = l.drop(columns=['ofz_ger_5', 'ofz_ust_5','ofz_ust_10'])
coint_test = coint_johansen(np.log(test_data),
                           det_order=0,
                           k_ar_diff=1)

summary = JohansenResults(coint_test)
summary.max_eigenvalue
summary.trace
```

Что это значит? В `max_eigenvalue` тесте (первая табличка) для гипотезы об одном коинтеграционном соотношении наблюдаемая статстика 40, а критическая на уровне 1% 39. Это значит, что на любом разумном уровне значимости не отвергается гипотеза об одном коинтеграционном соотношении в системе.

Но если мы оставим в системе два спреда, то они будут коинтегрированы между собой, так как ставки в США и ЕС взаимосвязаны, и мы получим два коинтеграционных вектора в системе:

```python
coint_test = coint_johansen(np.log(l.drop(columns=['ofz_ger_5', 
                                                   'ofz_ust_5',
                                                   'ofz_ust_10',
                                                   'ofz_ger_10'])),
                           det_order=0,
                           k_ar_diff=1)

summary = JohansenResults(coint_test)
# H0: Rank = 2. Observed statistics: 49, critical: 39 (1%). H0 not rejected.
# H0: Rank = 3. Observed statistics: 18.52, critical: 25.12 (10%). H0 rejected.
summary.max_eigenvalue
summary.trace
```

Нам нужно с какими-то весами взвесить спреды доходностей ОФЗ к UST и Bunds. Почему? Потому что между ставками в США и ЕС существует коинтеграция, которую легко объяснить теоретически. Из-за этого в системе появляется второе, не нужное нам коинтеграционное соотношения. Либо мы можем просто оставить какой-то из двух спредов, поскольку эти два спреда коинтегрированы между собой.

Я решил присвоить спредам веса по 0.5 и создать синтетическую переменную, так как фондовый рынок США - крупнейший в мире, и, безусловно, инвесторы по всему миру ориентируются на динамику доходностей Treasuries, а также решения ФРС по ставке. С другой стороны, ЕС - крупнейший торговый партнер России. Ориентироваться будем на 10-летние облигации как на наиболее ликвидные.

```python
l['spread'] = l['ofz_ger_10']
```

```python
from statsmodels.tsa.stattools import adfuller


def adf_test(timeseries):
    dftest = adfuller(timeseries, autolag="AIC", regression='n')
    dfoutput = pd.Series(
        dftest[0:4],
        index=[
            "Test Statistic",
            "p-value",
            "#Lags Used",
            "Number of Observations Used",
        ],
    )
    for key, value in dftest[4].items():
        dfoutput["Critical Value (%s)" % key] = value
    return dfoutput
```

Впрочем, график логарифмов не очень информативен. Теперь перейдем собственно к модели.

Можно путем несложных алгебраических преобразований показать, что коинтегрированная система для переменных в уровнях, описанная через `VAR(p=n_lags)` сводится к виду `VECM(p-1)` для переменных в разностях:

$$
\begin{aligned}
    VAR(p): \; X_t & = \Pi_1 X_{t-1} + \ldots + \Pi_p X_{t-p} + \varepsilon_t \\
    VECM(p-1): \; \bigtriangleup X_t & = \Pi_1 X_{t-1} + \Gamma_1 \bigtriangleup X_{t-2} + \ldots + \Gamma_{p-1} \bigtriangleup X_{t-p+1} + \varepsilon_t
\end{aligned}
$$

Дальше можно показать, что регрессор $X_{t-1}$во втором уравнении можно заменить на остатки регрессии $X_t$ на $X_{t-1}$, и тогда $\Pi_1$ приобретет новый смысл: это будут те самые speed adjustment coefficients, коэффициенты скорости подстройки эндогенных переменных системы к долгосрочному равновесию.

Мы будем строить модель следующим образом:
1. Сначала получим time-varying оценки параметров для линейной регрессии в коинтеграционном соотношении для переменных в уровнях. В коинтеграционное соотношение войдут все переменные. Иначе говоря, мы оценим регрессию вида $X_t = \Pi_1 X_{t-1}$ и воспользуемся ее остатками, $\hat\varepsilon$.<br>
2. Потом возьмем остатки из пункта (1) и будем использовать их как регрессор (Error Correction Term) в VAR(1) модели. В VAR(1) все переменные будут преобразованы к формату лог-доходностей (поскольку разность логарифмов аналогична логарифму доходности).<br>
    * Экзогенными переменными будут курс евро и цена на нефть.<br>
    * Эндогенными будут курс рубля, индекс ММВБ и синтетический спред ОФЗ к UST и Bunds (10-летние бумаги).<br>
    
Начнем с time-varying parameter регрессии из пункта (1) для коинтеграционного соотношения. Модель будет выглядеть следующим образом:


$$
\begin{aligned}
    USDRUB_t = d + BRENT_t \beta_{b,t} + EU&RUSD_t \beta_{e,t} + IMOEX_t \beta_{i,t}+ SPREAD_t \beta_{s,t} + \varepsilon_t\\[5pt] 
    \varepsilon_t & \sim N(0, \sigma_\varepsilon^2) \\[7pt]
    \begin{bmatrix} 
        \beta_{b,t} \\
        \beta_{e,t} \\
        \beta_{i,t} \\
        \beta_{s,t} \\
     \end{bmatrix}  = 
     \begin{bmatrix} 
        \beta_{b,t-1} \\
        \beta_{e,t-1} \\
        \beta_{i,t-1} \\
        \beta_{s,t-1} \\
      \end{bmatrix} + 
      \begin{bmatrix} 
        \zeta_{b,t-1} \\
        \zeta_{e,t-1} \\
        \zeta_{i,t-1} \\
        \zeta_{s,t-1} \\
      \end{bmatrix} \hspace{2.8em} 
      & \begin{bmatrix} 
        \zeta_{b,t-1} \\
        \zeta_{e,t-1} \\
        \zeta_{i,t-1} \\
        \zeta_{s,t-1} \\
      \end{bmatrix} \sim N \left ( \begin{bmatrix} 
                                        0 \\ 0 \\ 0 \\ 0 
                                    \end{bmatrix}, 
                                    \begin{bmatrix} 
                                        \sigma_{\beta, b}^2 & 0 & 0 & 0 \\
                                        0 & \sigma_{\beta, e}^2 & 0 & 0 \\ 
                                        0 & 0 & \sigma_{\beta, i}^2 & 0 \\
                                        0 & 0 & 0 & \sigma_{\beta, s}^2 \\
                                    \end{bmatrix} \right )
\end{aligned}
$$

```python
from collections import OrderedDict
import statsmodels.api as sm
from typing import Union

class TVPRegression(sm.tsa.statespace.MLEModel):
    def __init__(self, 
                 y:Union[pd.Series, np.array], 
                 X:pd.DataFrame,
                 var_guess:list,
                 positive_params:Union[slice, bool]=None):
        # shaped nobs x k_states
        exog = X.values

        super(TVPRegression, self).__init__(
            endog=y, 
            exog=exog, 
            k_states=X.shape[1], 
            initialization="diffuse"
        )

        # Since the design matrix is time-varying, it must be
        # shaped k_endog x k_states x nobs
        # Notice that exog.T is shaped k_states x nobs, so we
        # just need to add a new first axis with shape 1
        self.ssm["design"] = exog.T[np.newaxis, :, :]  # shaped 1 x 4 x nobs
        self.ssm["selection"] = np.eye(self.k_states)
        self.ssm["transition"] = np.eye(self.k_states)
        # add some helper attributes
        self.exog_names = X.columns.tolist()
        self.var_guess = var_guess

        # Which parameters need to be positive?
        # Those from covariance matrices
        self.positive_parameters = positive_params if positive_params else \
        slice(1, self.k_states+2) 

    @property
    def param_names(self):
        return ["intercept", 
                "var.e",
                *[f"var.{name}.coeff" for name in self.exog_names]]


    @property
    def start_params(self):
        """
        Defines the starting values for the parameters
        The linear regression gives us reasonable starting values for the constant
        d and the variance of the epsilon error
        """
        exog = sm.add_constant(self.exog)
        fitted = sm.OLS(self.endog, exog).fit()
        params = np.array([fitted.params[0], 
                           fitted.scale, 
                           *self.var_guess]).ravel()
        return params

    def transform_params(self, unconstrained):
        """
        We constraint the last k_states + 1 parameters
        ('var.e', 'var.{...}.coeff', 'var.{...}.coeff, ...') 
        to be positive, because they are variances
        """
        constrained = unconstrained.copy()
        constrained[self.positive_parameters] = (
            constrained[self.positive_parameters] ** 2
        )
        return constrained

    def untransform_params(self, constrained):
        """
        Need to unstransform all the parameters you transformed
        in the `transform_params` function
        """
        unconstrained = constrained.copy()
        unconstrained[self.positive_parameters] = (
            unconstrained[self.positive_parameters] ** 0.5
        )
        return unconstrained

    def update(self, params, **kwargs):
        params = super(TVPRegression, self).update(params, **kwargs)

        self["obs_intercept", 0, 0] = params[0]
        self["obs_cov", 0, 0] = params[1]
        self["state_cov"] = np.diag(params[2:(self.k_states+2)])
```

Оценим модель VECM.

```python
class TVPVAR(sm.tsa.statespace.MLEModel):
    # шаги 2-3 будем делать внутри конструктора класса
    def __init__(self, y, X):
        # матрица [y_t' : y_{t-1}'] для t = 2, ..., T
        augmented = sm.tsa.lagmat(y, 1, trim='both', original='in', use_pandas=True)
        # разделим на y_t и z_t = [1 : y_{t-1}']
        p = y.shape[1]
        y_t = augmented.iloc[:, :p]
        print(augmented)
        z_t = np.concatenate((X, augmented.iloc[:, p:]), axis=1)
        
        # Вспомним, что длина вектора состояния равна p * (p + 1)
        k_states = p * (p + 1)
        super().__init__(y_t, exog=z_t, k_states=k_states)
        
        # по умолчанию матрицы модели "состояние-наблюдение" заполнены нулями
        # поэтому нет нужды руками прописывать, что c_t = d_t = 0
        
        # теперь зададим матрицу Z_t
        # -> self.k_endog = p размерность наблюдаемого вектора
        # -> self.k_states = p * (p + 1) размерность преобразованного вектора коэфф-тов
        # -> self.nobs = T число наблюдений y_t
        
        # создаем список из k_endog матриц, 
        # каждая из которых имеет размерность k_states строк х n_obs столбцов
        # то есть размерность Z_t (4, 20, 201)
        # размерность каждой матрицы (20, 201)
        self['design'] = np.zeros((self.k_endog, self.k_states, self.nobs))
        for i in range(self.k_endog):
            start = i * (self.k_endog + 1)
            end = start + self.k_endog + 1
            # i = 0 
            # -> start = 0, end = 5
            # self['design', 0, 0:5, :] = z_t.T
            # то есть в первую матрицу в строки 0-5 записываем z_t^T (5 x 201)
            # i = 1
            # -> start = 5, end = 10
            # self['design', 1, 5:10, :] = z_t.T
            # то есть во вторую матрицу в строки 5-10 записываем транспонированную z_t
            self['design', i, start:end, :] = z_t.T
            
        # зададим матрицу T = I
        self['transition'] = np.eye(k_states)
        
        # зададим матрицу R = I
        self['selection'] = np.eye(k_states)
        
        # Шаг 3: инициализируем вектор состояния как alpha_1 ~ N(0, 5I)
        self.ssm.initialize('known', stationary_cov=5 * np.eye(self.k_states))
        
    # Шаг 4: создадим метод для обновления ковариационных матриц H и Q
    def update_variances(self, obs_cov, state_cov_diag):
        self['obs_cov']  = obs_cov
        self['state_cov'] = np.diag(state_cov_diag)
        
    # наконец, определим человекочитаемые имена для каждого элемента
    # вектора состояния
    @property
    def state_names(self):
        state_names = np.empty((self.k_endog, self.k_endog + 1), dtype=object)
        for i in range(self.k_endog):
            endog_name = self.endog_names[i]
            state_names[i] = (
            [f"speed_adj.{endog_name}"] +
            [f"L1.{other_name}->{endog_name}" for other_name in self.endog_names]
            )
            
        return state_names.ravel().tolist()
        
```

```python
def plot_coefficients_by_equation(states):
    fig, axes = plt.subplots(3, 1, figsize=(15, 24))
    
    # мы определили Z_t так, что первые 5 элементов вектора состояния
    # относятся к первой переменной в y_t, то есть GDP growth
    ax = axes[0]
    states.iloc[:, :4].plot(ax=ax)
    ax.set_title('USDRUB_TOM', y=1.02)
    ax.set_xlabel('Дата')
    ax.legend(fontsize=16)
    ax.grid(True)
    
    ax = axes[ 1]
    states.iloc[:, 4:8].plot(ax=ax)
    ax.set_title("IMOEX", y=1.02)
    ax.set_xlabel('Дата')
    ax.legend(fontsize=16)
    ax.grid(True)
    
    ax = axes[2]
    states.iloc[:, 8:12].plot(ax=ax)
    ax.set_title('SPREAD', y=1.02)
    ax.set_xlabel('Дата')
    ax.legend(fontsize=16)
    ax.grid(True)
    
    fig.tight_layout()
    fig.savefig(image_dir/'tvpvar.eps', dpi=500)
    
    return ax
```

```python
# априорные гиперпараметры

# для H априорное распределение inverse-Wishart(v_1^0=k + 3, S10=I)
v10 = mod.k_endog + 2
S10 = np.eye(mod.k_endog)

# для ков. матрицы коэффициентов априорное 
# inverse-Gamma(v_{i2}^0 / 2 = 3, S+{i2}^0 / 2 = 0.005)
vi20 = 4
Si20 = 0.01
```

```python
# настройка сэмплера Гиббса
niter = 1000
nburn = 100

# 1. массивы для хранения информации
store_states = np.zeros((niter + 1, mod.nobs, mod.k_states))
store_obs_cov = np.zeros((niter + 1, mod.k_endog, mod.k_endog))
store_state_cov = np.zeros((niter + 1, mod.k_states))

# 2. Загрузим стартовые занчения
store_obs_cov[0] = initial_obs_cov
store_state_cov[0] = initial_state_cov_diag
mod.update_variances(store_obs_cov[0], store_state_cov[0])

# 3. Инициализируем сэмплировщик
sim = mod.simulation_smoother(method='cfa')
```

```python
from scipy.stats import invwishart, invgamma
```

```python
%%time
for i in range(niter):
    mod.update_variances(store_obs_cov[i], store_state_cov[i])
    sim.simulate()
    
    # 1. Сэмплируем состояния
    store_states[i+1] = sim.simulated_state.T
    
    # симулируем ковариационную матрицу наблюдений
    fitted = np.matmul(mod['design'].transpose(2, 0, 1),
                       store_states[i + 1][..., None])[..., 0]
    resid = mod.endog - fitted
    store_obs_cov[i+1] = invwishart.rvs(v10 + mod.nobs, S10 + resid.T @ resid)
    
    # симулируем ковариационную матрицу вектора состояний
    resid = store_states[i + 1, 1:] - store_states[i + 1, :-1]
    sse = np.sum(resid**2, axis=0)
    
    for j in range(mod.k_states):
        rv = invgamma.rvs((vi20 + mod.nobs - 1) / 2, scale=(Si20 + sse[j]) / 2)
        store_state_cov[i + 1, j] = rv
```

```python
def rw_forecast(y:pd.DataFrame):
    forecast = y.shift(1).dropna()
    error = y.iloc[1:] - forecast
    return np.mean(error **2)
```

```python
def TVP_forecast(lm_y,
                 lm_X,
                 y,
                 niter, 
                 nburn):
    
    mod = TVPRegression(lm_y, lm_X, 
                        var_guess=[0.01 for _ in range(4)]
                       )
    
    res = mod.fit()
    residuals = lm_y - res.fittedvalues
    residuals = pd.Series(data=residuals,
                          index=lm_X.index)
    
    X = residuals.iloc[1:-1].values.reshape((-1, 1))

    mod = TVPVAR(y, X)
    
    v10 = mod.k_endog + 2
    S10 = np.eye(mod.k_endog)
    vi20 = 4
    Si20 = 0.01
    
    store_states = np.zeros((niter + 1, mod.nobs, mod.k_states))
    store_obs_cov = np.zeros((niter + 1, mod.k_endog, mod.k_endog))
    store_state_cov = np.zeros((niter + 1, mod.k_states))
    
    store_obs_cov[0] = initial_obs_cov
    store_state_cov[0] = initial_state_cov_diag
    mod.update_variances(store_obs_cov[0], store_state_cov[0])
    
    sim = mod.simulation_smoother(method='cfa')
    
    for i in range(niter):
        
        mod.update_variances(store_obs_cov[i], store_state_cov[i])
        sim.simulate()
        
        store_states[i+1] = sim.simulated_state.T
        
        fitted = np.matmul(mod['design'].transpose(2, 0, 1),
                           store_states[i + 1][..., None])[..., 0]
        resid = mod.endog - fitted
        store_obs_cov[i+1] = invwishart.rvs(v10 + mod.nobs, S10 + resid.T @ resid)
        
        resid = store_states[i + 1, 1:] - store_states[i + 1, :-1]
        sse = np.sum(resid**2, axis=0)
        
        for j in range(mod.k_states):
            rv = invgamma.rvs((vi20 + mod.nobs - 1) / 2, scale=(Si20 + sse[j]) / 2)
            store_state_cov[i + 1, j] = rv
        
    state_posterior_mean = pd.DataFrame(
    np.mean(store_states[nburn + 1:], axis=0),
    index = y.index[1:], columns = mod.state_names
    )
    
    # forecast
    Gamma_t = state_posterior_mean.iloc[-1, :].values.reshape(-1, 
                                                         y.shape[1]+1)
    reg = y.iloc[-1].values.reshape((y.shape[1], -1))
    reg = np.concatenate((X[-1].reshape(-1, 1), reg), axis=0)
    f = Gamma_t.dot(reg)
    
    return f
    
```

```python
import pymc3 as pm
import theano
import theano.tensor as tt
```

```python
class Loglike(tt.Op):

    itypes = [tt.dvector]  # expects a vector of parameter values when called
    otypes = [tt.dscalar]  # outputs a single scalar value (the log likelihood)

    def __init__(self, model):
        self.model = model
        self.score = Score(self.model)

    def perform(self, node, inputs, outputs):
        (theta,) = inputs  # contains the vector of parameters
        llf = self.model.loglike(theta)
        outputs[0][0] = np.array(llf)  # output the log-likelihood

    def grad(self, inputs, g):
        # the method that calculates the gradients - it actually returns the
        # vector-Jacobian product - g[0] is a vector of parameter values
        (theta,) = inputs  # our parameters
        out = [g[0] * self.score(theta)]
        return out


class Score(tt.Op):
    itypes = [tt.dvector]
    otypes = [tt.dvector]

    def __init__(self, model):
        self.model = model

    def perform(self, node, inputs, outputs):
        (theta,) = inputs
        outputs[0][0] = self.model.score(theta)
```

```python
# Set sampling params
ndraws = 1e3  #  1e3 number of draws from the distribution
nburn = 1e2  # 600 number of "burn-in points" (which will be discarded)
```

```python
# Construct an instance of the Theano wrapper defined above, which
# will allow PyMC3 to compute the likelihood and Jacobian in a way
# that it can make use of. Here we are using the same model instance
# created earlier for MLE analysis (we could also create a new model
# instance if we preferred)
loglike = Loglike(mod)

with pm.Model():
    # Priors
    intercept = pm.Uniform("intercept", 0, 5)
    var_e = pm.InverseGamma("var.e", 2.3, 0.5)
    var_b_coeff = pm.InverseGamma("var.b.coeff", 2.3, 0.1)
    var_e_coeff = pm.InverseGamma("var.e.coeff", 2.3, 0.1)
    var_i_coeff = pm.InverseGamma("var.i.coeff", 2.3, 0.1)
    var_s_coeff = pm.InverseGamma("var.s.coeff", 2.3, 0.1)

    # convert variables to tensor vectors
    theta = tt.as_tensor_variable([intercept, 
                                   var_e, 
                                   var_b_coeff, 
                                   var_e_coeff,
                                   var_i_coeff,
                                   var_s_coeff])

    # use a DensityDist (use a lamdba function to "call" the Op)
    pm.DensityDist("likelihood", loglike, observed=theta)

    # Draw samples
    trace = pm.sample(
        ndraws,
        tune=nburn,
        return_inferencedata=True,
        cores=1,
        compute_convergence_checks=False
    )
```
